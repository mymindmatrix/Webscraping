```python
#importing necessary library
import requests
import bs4
from bs4 import BeautifulSoup as bs
from time import sleep
from random import randint
```

```python
#creating list to store data from webscraping
products = []
prices = []
ratings = []
no_of_ratings = []
ram_rom = []
screen = []
camera=[]
battery = []
processor = []
```

```python
# extracting mobile data from flipkart using loop from multiple page
No_of_pages=5
for i in range(1,int(No_of_pages)+1):
    url='https://www.flipkart.com/search?q=smartphone&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_3_3_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_3_3_na_na_na&as-pos=3&as-type=RECENT&suggestionId=smartphone&requestId=f81ae89b-1daa-4f14-820f-882a78e48350&as-searchtext=sma&page='+str(i)
    response=requests.get(url)
    content=bs(response.content, 'html.parser')
    sleep(randint(2,8))
    name=content.find_all('div',{'class':'_4rR01T'})
    price=content.find_all('div',{'class':'_30jeq3 _1_WHN1'})
    rate=content.find_all('span',{'class':'_1lRcqv'})
    no_of_rate=content.find_all('span',{'class':'_2_R_DZ'})
    specification=content.find_all('div',{'class':'fMghEO'})
    
    
    for i in name:
        products.append(i.text)
    for i in price:
        prices.append(i.text)
    for i in rate:
        ratings.append(i.text)
    for i in no_of_rate:
        no_of_ratings.append(i.text)
    for col in specification:
        col=col.find_all('li',{'class':'rgWa7D'})
        rr=col[0].text
        ram_rom.append(rr)
        disp=col[1].text
        screen.append(disp)
        cam=col[2].text
        camera.append(cam)
        bat=col[3].text
        battery.append(bat)
        proc=col[4].text
        processor.append(proc)
    
```

```python
import pandas as pd # importing pandas to create dataframe
```

```python
#checking the output numbers
print(len(products))
print(len(prices))
print(len(ratings))
print(len(no_of_ratings))
print(len(ram_rom))
print(len(screen))
print(len(camera))
print(len(battery))
print(len(processor))
```

```python
data=pd.DataFrame({'Smartphone Name':products, 'Price':prices, 'Ram & Rom':ram_rom, 'Dipslay':screen, 'Camera':camera, 'Battery Capacity':battery, 'Operating System':processor, 'Ratings':ratings, 'No of Ratings': no_of_ratings})
```

```python
data.head()
```

```python
data=data.rename(columns={'Ratings':'Star_Ratings'})
data=data.rename(columns={'No of Ratings':'rate'})
```

```python
data[['No of Rate', 'Ratings','&', 'No of Reviews', 'Revies']]=data.rate.str.split(expand=True)
```

```python
data['Price']=data['Price'].str.lstrip('₹') # removing symobol to convert string into number
```

```python
data.head()
```

```python
data.drop(['rate', 'Ratings','&','Revies'],axis=1, inplace=True)
```

```python
data['Price']=data['Price'].replace({',':''},regex=True)
data['No of Rate']=data['No of Rate'].replace({',':''},regex=True)
data['No of Reviews']=data['No of Reviews'].replace({',':''},regex=True)
```

```python
data.info()
```

```python
#converting price and rating into float
data['Price']=data['Price'].astype(float)
data['Star_Ratings']=data['Star_Ratings'].astype(float)
data['No of Rate']=data['No of Rate'].astype(int)
data['No of Reviews']=data['No of Reviews'].astype(int)
```

```python
data=data.rename(columns={'No of Rate':'No of Ratings'}) 
```
